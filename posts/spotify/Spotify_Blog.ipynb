```python
import pandas as pd
```

```python
spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')
spotify
```

I really enjoy the spotify playlist "Throwbacks".

```python
throwbacks = spotify['playlist_name'] == "Throwbacks"
throwbacks_df = spotify[throwbacks]
throwbacks_df
```

```python
throwbacks_df['artist_name'].nunique()
```

There is a total of 223 different artists in this playlist, including one of my favorite artists, Britney Spears.

```python
by_artist = throwbacks_df.set_index(keys = "artist_name")
by_artist
```

```python
britney = throwbacks_df['artist_name'] == "Britney Spears"
britney_df = throwbacks_df[britney]
britney_df
```

```python
britney_df['track_name'].count()
```

She has 16 different songs in the playlist:

```python
britney_df['track_name']
```

```python
song_length = britney_df['duration_ms'].sort_values(ascending=False)
song_length
song_length_min = song_length/60000
song_length_min
```

Her longest song is 4.187 minutes, while her shortest song is 3.206 minutes.

```python
song_length_min.sum()
```
